```python
import asyncio
import yaml

# EmbedChain App class
from embedchain.apps.open_source_app import App

# Configuration string
config = """
llm:
  provider: gpt4all
  config:
    model: 'orca-mini-3b.ggmlv3.q4_0.bin'
    temperature: 0.5
    max_tokens: 1000
    top_p: 1
    stream: false

embedder:
  provider: gpt4all
  config:
    model: 'all-MiniLM-L6-v2'
"""

# Write the multi-line string to a YAML file
with open('gpt4all.yaml', 'w') as file:
    file.write(config)

class MyApp:
    def __init__(self, app):
        self.app = app
        self.links = []  # Initialize the links attribute

    async def add(self, link):
        # Simulate adding a link and processing it asynchronously
        print(f"Adding link: {link}")
        
        # Add the link to the app
        self.app.add(link)
        
        # Append to local links list
        self.links.append(link)
        print(f"Link added: {link}")

    async def query(self, question):
        # Query the app
        answer = self.app.query(question)
        return answer

async def main():
    # Initialize the embedchain App from the YAML configuration
    embedchain_app = App.from_config(yaml_path="gpt4all.yaml")
    
    # Pass the initialized embedchain App to our custom MyApp class
    my_app = MyApp(embedchain_app)
    
    while True:
        print("\nChoose an option:")
        print("1. add")
        print("2. ask")
        print("3. quit")
        
        choice = input("Enter choice: ")

        if choice == "1":
            link = input("Enter link: ")
            await my_app.add(link)
        elif choice == "2":
            question = input("Enter question: ")
            answer = await my_app.query(question)
            print(answer)
        elif choice == "3":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please choose again.")


# Run the asynchronous main function in Jupyter or IPython
await main()
```
